```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("diabetes.csv")
data
```

```python
X=data.drop(["Outcome"],axis=1)
y=data["Outcome"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
# create decisiontree classfier object
cls=DecisionTreeClassifier()

cls.fit(X_train,y_train)
```

```python
score = cls.score(X_train,y_train)
print("R-squared Train  :", score)

score = cls.score(X_test, y_test)
print("R-squared Test  :", score)

y_pred = cls.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",mse**(1/2.0))
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))

tree.plot_tree(cls,filled=True)
plt.show()
```

```python
cls=DecisionTreeClassifier(criterion="entropy",max_depth=4)

cls.fit(X_train,y_train)

score = cls.score(X_train,y_train)
print("R-squared Train  :", score)

score = cls.score(X_test, y_test)
print("R-squared Test  :", score)

y_pred = cls.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",mse**(1/2.0))
```

```python
from sklearn import tree

plt.figure(figsize=(20,15))

tree.plot_tree(cls,filled=True,class_names=True)
plt.show()
```
